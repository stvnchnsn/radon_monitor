## Motivations <a name = 'motivations'></a>

[Radon](https://www.cdc.gov/nceh/features/protect-home-radon/index.html) is a radioactive gas that forms naturally when uranium, thorium, or radium, which are radioactive metals break down in rocks, soil and groundwater.  It is the second leading cause of lung cancer.  The CDC recommends mitigation if radon levels exceed 4 picocuries per liter (pCi/L).  


My basement was professionally tested in the winter of 2016 and was measured at around 3.5 pCi/L.  At the time we considered this acceptable because the basement would mainly be used for storage.  However, during the initial months of the pandemic and the normalization of work from home I started to spend more time in the basement working as it was the quietest place in the house.  This lead me to purchase an [AirThings](https://www.amazon.com/Corentium-Detector-Airthings-223-Lightweight/dp/B00H2VOSP8/ref=asc_df_B00H2VOSP8/?tag=hyprod-20&linkCode=df0&hvadid=312148082093&hvpos=&hvnetw=g&hvrand=3126199023951679431&hvpone=&hvptwo=&hvqmt=&hvdev=c&hvdvcmdl=&hvlocint=&hvlocphy=9021463&hvtargid=pla-568997698229&psc=1) radon detector.  After installing the detector it became apparent overtime from ancedotal observations over several months that the levels could vary wildly. This observation motivated more research and after reading about potential correlation between weather conditions and radon level fluctuations as well as the varying levels a radon mitigation I decided to build a system to track radon measurements over time to compare the radon measurements to weather conditions so that the effectiveness of any radon mitigation efforts could be evaluated accuratly. 

## System Overview <a name = 'system_overview'></a>
[Table of Contents](#toc)

A Raspberry Pi Zero (referred to as camerapi here) is mounted to take pictures of the AirThings Radon Monitor Display.  The camerapi takes a picture every hour and uploads the image to a Raspberry Pi 4 running a Samba file server.  The AirThings display always displays a long-term average but fluctuates between displaying a daily average and a 7-day average; therfore, it's somewhat random which short-term time frame is being displayed.  However, taking a picture every hour means that enough 1-day measurements are taken to be useful.  

Local weather data is pulled hourly using the same Raspberry Pi 4 from openweather.com.  The weather data is stored in a database.

The images can then be read using two machine learning models (neural net and logistic regression) as an ensemble that were developed using the model.ipynb notebook.  In the future this functionality will be rolled into the Radon_Monitor class.  Two models are used because occasionally the picture is of a 'transition' meaning the display is transitioning from one mode to the next as well as to increase confidence that the readings are correct.  Images where the two models disagree are not used; however, they can be reviewed.

### Model Training

Intially a logical regression model was used when the training set was low as more data was gathered and after balancing and artifically increasing the dataset a neural network model was developed.  The logistic regression model tends to overfit the training data and perform not as well on new examples; however, it is still useful because where the models agree is 99% accurate.  Therfore, the final model is an ensemble of the logistic regression and the neural network model.  
In the future the plan is to improve the neural network model by using methods like distoring the training data as well as using a moving window as oppose to the manual cropping/calibration currently being used.

```python
import sqlite3
conn = sqlite3.connect('/Volumes/sambashare/projects/radon_monitor/data/environmental_data.sqlite')
cursor = conn.cursor()
q = '''
SELECT date, temp,pressure,humidity
FROM daily_weather'''
cursor.execute(q)
rows= cursor.fetchall()
daily_weather_df = pd.DataFrame.from_records(rows,columns = ['DATE','hourly_temp','pressure','humidity'])
daily_weather_df['DATE'] = pd.to_datetime(daily_weather_df['DATE'])
daily_weather_df.tail()
```

```python
# weather_data = pd.read_csv('./reference_data/clean_weather_data.csv')
fig, ax1 = plt.subplots(figsize = (15,5))
weather_data["DATE"] = pd.to_datetime(weather_data["DATE"])
weather_data = weather_data[['DATE','TMAX','TMIN']][weather_data["DATE"]>pd.to_datetime('11/01/2021')]

weather_data = pd.concat([weather_data,daily_weather_df])


x = models_agree_df.img_date[models_agree_df['nn_short_term_span'] == 1]
x= pd.to_datetime(x)
y = models_agree_df.nn_short_term[models_agree_df['short_term_span']==1]

ax2 = ax1.twinx()
ax1.plot(weather_data['DATE'], weather_data['TMAX'], 'g-',label = 'TMAX')
ax1.plot(weather_data['DATE'], weather_data['TMIN'], 'c-',label = 'TMIN')
ax1.plot(weather_data['DATE'], weather_data['hourly_temp'], 'g--',label = 'hourly_temp')
ax2.plot(x, y, 'kx',label='Radon')
ax2.plot(x,len(x)*[4],color = 'k',label='Radon 4.0')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature')
ax2.set_ylabel('Radon')
ax1.grid()
fig.legend(loc = 'upper right')
plt.savefig('./figures/results.png')
plt.show()
```

### Correlation Between Environmental Readings and Radon Readings

```python
# for correlation only hourly temp is used
weather_df = weather_data.dropna(subset = ['hourly_temp'])\
                        .reset_index(drop = True)[['DATE','hourly_temp','pressure','humidity']]
weather_df['date'] =  weather_df.DATE.dt.date 
weather_df['hour'] = weather_df.DATE.dt.hour
weather_df.head()
```

```python
radon_df = models_agree_df[['img_date','short_term']][models_agree_df['short_term_span']==1]
radon_df.img_date = pd.to_datetime(radon_df.img_date)
radon_df['date'] = radon_df.img_date.dt.date
radon_df['hour'] = radon_df.img_date.dt.hour
radon_df.head()
```

```python
combined_df = weather_df.merge(radon_df, on = ['date','hour'],how = 'left').dropna(subset = ['short_term']).reset_index(drop = True)
combined_df.head()
```

```python
f,axes = plt.subplots(nrows = 3,figsize = (10,10))
axes[0].scatter(combined_df['hourly_temp'],combined_df['short_term'])
axes[0].set_xlabel('hourly_temp')
axes[1].scatter(combined_df['humidity'],combined_df['short_term'])
axes[1].set_xlabel('humdity')
axes[2].scatter(combined_df['pressure'],combined_df['short_term'])
axes[2].set_xlabel('pressure')
f.suptitle('Environmental Measures vs Radon Readings')
plt.show()
```
