```python
# Raw Files
fp = '/Volumes/wellington/projects/radon_monitor/data/hourly_tests/'
photo_files = [i.name for i in os.scandir(fp)]
photo_path = [fp+fn for fn in photo_files]
photo_dates = [date_extractor(fp+fn) for fn in photo_files]
print('Number of files found = {}'.format(len(photo_path)))

date_path_map = pd.DataFrame()
for i, (date, path) in enumerate(zip(photo_dates,photo_path)):
    date_path_map.loc[i,'date'] = date
    date_path_map.loc[i,'path'] = path
date_path_map = date_path_map.sort_values(by = 'date').reset_index(drop = True)
date_path_map.head()
```
